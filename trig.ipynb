```python
x = sp.symbols('x')
sin_expr = sp.sin(x)
cos_expr = sp.cos(x)
display(Markdown(f"**Sine Expression:** {sp.latex(sin_expr)}"))
display(Markdown(f"**Cosine Expression:** {sp.latex(cos_expr)}"))
```

```python
# Create a rotation matrix around the z-axis
x = sp.symbols('x')
rotation_matrix = sp.Matrix([[sp.cos(x), -sp.sin(x), 0],
                             [sp.sin(x), sp.cos(x), 0],
                             [0, 0, 1]])

# Create a translational matrix that moves the point to the origin
translation_matrix = sp.Matrix([[x, 0, 0],
                                [0, x, 0],
                                [-1, -1, x]])

composite_matrix = rotation_matrix * translation_matrix
composite_matrix
```

```python
composite_matrix.diff(x)
```

```python
composite_matrix.integrate(x)
```

```python
composite_matrix.subs(x, 1).evalf()
```
